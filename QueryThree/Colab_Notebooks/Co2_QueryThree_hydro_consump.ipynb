```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'https://raw.githubusercontent.com/giofile/Project_CO2_Emissions/main/QueryThree/Data_Sources/hydropower-consumption.csv'
data = pd.read_csv(file_path).copy()

# Calculate the average hydropower consumption for each entity
average_consumption = data.groupby('Entity')['Electricity from hydro - TWh'].mean().copy()

# Sort the entities based on average consumption and pick the top 5
top_entities = average_consumption.sort_values(ascending=False).head(5).index.copy()

# Filter the data for these top entities
data_top = data[data['Entity'].isin(top_entities)].copy()

# Plotting
plt.figure(figsize=(10, 6))
for entity in top_entities:
    entity_data = data_top[data_top['Entity'] == entity].copy()
    plt.plot(entity_data['Year'], entity_data['Electricity from hydro - TWh'], label=entity)

plt.title('Top Entities by Average Hydropower Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Electricity from Hydro (TWh)')
plt.legend(title='Entity/Region')
plt.grid(True)
plt.show()
```

**Hydropower Consumption Trends Analysis Conclusion:**

The analysis of "Top Entities by Average Hydropower Consumption Over Time" demonstrates a global increase in the adoption of hydropower, with notable growth in both G20 and World categories, **reflecting broad investments across developed and developing economies**. The recent stabilization in consumption trends suggests potential market saturation or shifts towards other renewable resources. Non-OECD countries exhibit faster growth in hydropower usage, likely driven by their expanding energy needs and untapped natural resources. These observations provide essential insights for policymakers and investors, indicating the diverse strategic approaches to hydropower development across different global regions.
