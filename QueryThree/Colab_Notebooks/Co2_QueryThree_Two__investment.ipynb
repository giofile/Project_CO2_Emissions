Introduction:

*This analysis examines investment trends in various renewable energy technologies over the period 2004-2018. By visualizing investment fluctuations, we aim to identify key patterns and growth trajectories across different sectors within the renewable energy industry.*


```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
def load_data(file_path):
    return pd.read_csv(file_path).copy()


# Load the investment data
investment_data_path = 'https://raw.githubusercontent.com/giofile/Project_CO2_Emissions/main/QueryThree/Data_Sources/investment-in-renewable-energy-by-technology.csv'
investment_data = load_data(investment_data_path)

# Check the first and last few rows to understand the structure
print(investment_data.head())
print(investment_data.tail())
```

```python
import pandas as pd

# Convert 'Year' to a datetime object
investment_data['Year'] = pd.to_datetime(investment_data['Year'], format='%Y').copy()

# Check data types and look for any missing values
print(investment_data.info())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))

# Plotting investment trends for each technology
technologies = ['Marine energy', 'Small hydropower', 'Geothermal energy', 'Biofuels', 'Biomass and waste-to-energy', 'Wind energy', 'Solar energy']
for tech in technologies:
    subset = investment_data[['Year', tech]].copy()
    plt.plot(subset['Year'], subset[tech], label=tech, marker='o')

plt.title('Investment Trends in Renewable Energy Technologies (2004-2018)')
plt.xlabel('Year')
plt.ylabel('Investment ($)')
plt.legend(title='Technology')
plt.grid(True)
plt.show()
```

**Conclusion**

The graph detailing investment trends in renewable energy technologies up to 2018 illustrates a significant increase in funding for solar and wind energy, which highlights a strong market preference and growth in these sectors. Investments in solar energy, in particular, show a sharp rise, underscoring the technology's escalating affordability and efficiency. Meanwhile, technologies like biofuels and geothermal energy demonstrate modest growth, indicating sustained but more conservative investment interest. Overall, this graph portrays a clear focus on solar and wind energy within the renewable sector, driven by their potential for scalability and cost-effectiveness.
