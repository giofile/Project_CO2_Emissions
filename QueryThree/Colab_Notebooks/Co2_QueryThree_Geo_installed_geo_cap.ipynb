```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/giofile/Project_CO2_Emissions/main/QueryThree/Data_Sources/installed-geothermal-capacity.csv')

# Convert 'Year' to datetime for better plotting
data['Year'] = pd.to_datetime(data['Year'], format='%Y').copy()

# Select a few key entities known for geothermal energy use
selected_entities = ['United States', 'Indonesia', 'Iceland', 'Philippines', 'Kenya']
filtered_data = data[data['Entity'].isin(selected_entities)].copy()

# Plotting the installed geothermal capacity over time
plt.figure(figsize=(10, 6))
for label, df in filtered_data.groupby('Entity'):
    plt.plot(df['Year'].dt.year, df['Geothermal energy capacity - MW'], label=label)

plt.yscale('log')  # Using logarithmic scale due to wide range of capacities
plt.title('Installed Geothermal Capacity Over Time for Key Countries')
plt.xlabel('Year')
plt.ylabel('Installed Capacity (MW)')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

**Conclusion: Trends in Installed Geothermal Capacity Across Key Countries**

The graph illustrates significant growth in installed geothermal capacity in countries like Indonesia and Kenya, showcasing aggressive expansions since 2000. Meanwhile, nations like Iceland and the Philippines have maintained steady growth, leveraging their natural geothermal resources. The United States shows a consistent but modest increase, indicating a stable but less aggressive approach to geothermal expansion compared to the other featured countries. These trends underscore the varying commitment levels to geothermal energy, which are influenced by geographical advantages, economic policies, and technological advancements in each country.
