```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projects/Co2-QueryThree/share-of-electricity-production-from-renewable-sources.csv')

# Convert 'Year' to datetime for better plotting
data['Year'] = pd.to_datetime(data['Year'], format='%Y').copy()

# Filter for a few key entities for clarity
selected_entities = ['United States', 'Germany', 'China', 'India', 'Brazil']
filtered_data = data[data['Entity'].isin(selected_entities)].copy()

# Plotting the share of electricity production from renewable sources
plt.figure(figsize=(10, 6))
for label, df in filtered_data.groupby('Entity'):
    plt.plot(df['Year'], df['Renewables - % electricity'], label=label)

plt.title('Share of Electricity Production from Renewable Sources Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.legend(title='Entity')
plt.grid(True)
plt.show()

```

**Conclusion: Renewable Electricity Production Trends**

The graph demonstrates a notable increase in the share of electricity generated from renewable sources in countries like Germany, China, and India, showcasing significant growth since the early 2000s. Brazil maintains a high percentage of renewables in its energy mix, reflecting its long-standing commitment to hydroelectric and other renewable sources. In contrast, the United States shows a more gradual increase, though the trend has become more pronounced in recent years. These trends indicate a global shift towards renewable energy, driven by technological advancements, policy initiatives, and growing environmental awareness, though the pace and extent of adoption vary by country.
