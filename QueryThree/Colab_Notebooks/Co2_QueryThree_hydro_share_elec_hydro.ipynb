*This analysis delves into the trends of hydroelectric power utilization across key global players, focusing on its role in the overall energy mix. By analyzing historical data, we aim to uncover patterns in hydroelectric energy production and consumption, offering insights into how different countries have sustained or shifted their reliance on this renewable resource. This exploration highlights the percentage share of electricity from hydro sources, with particular emphasis on significant hydroelectric producers such as Brazil, Canada, and Norway.*

```python
import pandas as pd

# load dataset
file_path = 'https://raw.githubusercontent.com/giofile/Project_CO2_Emissions/main/QueryThree/Data_Sources/share-electricity-hydro.csv'
data = pd.read_csv(file_path).copy()

# Display the first and last few rows of the dataframe and the data types of the columns
print(data.head())
print(data.tail())
print(data.dtypes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'https://raw.githubusercontent.com/giofile/Project_CO2_Emissions/main/QueryThree/Data_Sources/share-electricity-hydro.csv'
data = pd.read_csv(file_path)

# Convert 'Year' to datetime for better plotting
data['Year'] = pd.to_datetime(data['Year'], format='%Y').copy()

# Select a few key entities known for significant hydroelectric production
selected_entities = ['Brazil', 'Canada', 'Norway', 'China', 'United States']
filtered_data = data[data['Entity'].isin(selected_entities)].copy()

# Plotting the share of electricity from hydro sources over time
plt.figure(figsize=(10, 6))
for label, df in filtered_data.groupby('Entity'):
    plt.plot(df['Year'].dt.year, df['Hydro - % electricity'], label=label, marker='o')

plt.title('Share of Electricity from Hydro Sources Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage of Electricity from Hydro (%)')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

**Conclusion of Hydro-Electric Power -Analysis**

The analysis of hydroelectric power trends across selected countries reveals that Brazil and Canada maintain a stable and high dependency on hydroelectric power, demonstrating its significant role in their energy mixes. In contrast, Norway shows a decline in hydroelectric power usage, possibly indicating a shift towards other renewable energy forms or a strategic reduction in hydro usage. The United States and China display variability in hydro utilization, with recent increases suggesting a renewed emphasis on enhancing hydroelectric capacity. Overall, the observed fluctuations and trends offer insights into how different nations adapt to and prioritize hydroelectric power within their broader energy strategies, reflecting a complex interplay of environmental, economic, and policy-driven factors.
